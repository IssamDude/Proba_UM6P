<h3> 1) Méthode de loi inverse : </h3>


Question <b>f</b> : $f_{X}(t)=\frac{1}{\pi(x²+1)}$

Réaliser une simulation de la variable aléatoire $X$.

Réponse : une primitive de $f_{X}(t)=\frac{1}{\pi(x²+1)}$ est : $F_{X}(t)=\frac{1}{2}+\frac{1}{\pi}\arctan{(t)}$.
$F$ réalise une bijection de $\mathbb{R}$ dans $]0,1[$.
Notons $G_{X}$ sa bijection réciproque. Un calcul simple permet d'obtenir $G_{X}(U)=\tan{(\pi U)}$, $U$ uniforme sur $]0,1[$.

$X$ et $G_{X}(U)$ suivent la même, nous allons donc générer X avec la méthode de l'inverse généralisée : 

```python
# Simulation de X par la méthode de l'inversé généralisée : 
import math
def simulX():
    return math.tan((math.pi)*np.random.random())

X = [simulX() for i in range(10000)]
plt.hist(X, range=(0,10),  bins=100) #edgecolor="red",
mn, mx = plt.xlim()
plt.xlim(0,mx)
intervallex=np.linspace(mn,mx,100)
def fX(x):
    return 1/((math.pi)*(1+x**2)) *1000
reelx=[fX(x) for x in intervallex]
plt.plot(intervallex, reelx )
```

```python
def uniform(a,b):
    return a+(b-a)*np.random.random()
```

```python
def fx(t):
    if 0<t<1 : 
        return t
    elif 1<=t<2:
        return 2-t
    else:
        return 0
```

```python
def gx(t,a,b) : 
    if a<t<b : 
        return 1/(b-a)
    else : 
        return 0
    
```

```python
t=[]
def acceptation_reget(a,b,c):
    while(len(t))<100000:
        j=0
        for i in range(1, 100000):
            u=uniform(a,b)
            v=np.random.random()
            n=fx(u)/gx(u,a,b)
            k=c*v
            if n> k :
                t.append(u)
            else : 
                j+=1
                continue
            #print(len(t))
        return plt.hist(t, range=(0,2), bins=50)
```

```python
def BoxMuller():
    u, v= np.random.random() , np.random.random()
    return math.cos(2*math.pi*u)*math.sqrt(-2*math.log(v))
```

<h4> a) En termes de complexité, comparer entre les deux méthodes de simulation de
la loi normale, centrale limite et box-Müller. </h4>

Bien que la transformation de Box-Muller a l'air d'avoir une compléxité moindre que la méthode du TCL, il est utile de remarquer que le TCL n'utilise que des opérations élémentaires ( somme & division ) tandis que B-M utilise des fonctions très coûteuses : cosinus et logarithme.
Cependant, pour que le TCL donne une génération fidèle à la loi normale, il faut un grand nombre n de générations d'uniformes, alors que la méthode Box-Muller n'éxige que 2 simulations d'unifromes.

<h4> b) Réaliser, via python, une simulation de la loi de khi-deux a n degrés de liberté
et tracer sa courbe de densité de probabilité (voir I.3.a). Penser à utiliser la
simulation de la loi normale réalisée selon la méthode de box-Müller : </h4>

```python
def BoxMuller2():
    u, v= np.random.random() , np.random.random()
    return math.cos(2*math.pi*u)*math.sqrt(-2*math.log(v)), math.sin(2*math.pi*u)*math.sqrt(-2*math.log(v))
def khi2(n) : #Simulation d'une loi du khi2 à n degrés de liberté
    h=0
    if n%2==0 :
        for i in range(int(n/2)) :
            v1, v2 = BoxMuller2()
            h+=(v1**2+v2**2)
    else : 
        for i in range(int(n-1/2)) : 
            v1, v2 = BoxMuller()
            h+=(v1**2+v2**2)
        h+=(BoxMuller2()[0])**2
    return h
            
        
        
```

```python
# tracé de la densité d'une loi du khi 2 à 2 degrés de liberés
def f_X(x, h):
    a=[]
    r=[khi2(2) for i in range(1000) ]
    for i in range (1000):
        
        if r[i]>x and r[i]<x+h : 
            a.append( r[i]/h)
    return len(a)/1000
        
```

```python
# Tracé de F_x sur [-2:2]
def f_X_I(intervalle) : 
    b=[]
    for i in range(len(intervalle)):
        b.append(10*f_X(intervalle[i], 0.1))
    return b
        
inte2 = np.arange(0,10, 0.01)

gg=f_X_I(inte2)
```

```python
def student(k) : # Loi de student à k degrés de libertés
    Z=BoxMuller()[0]
    K=math.sqrt(khi2(k)/k)
    return Z/K
```
